```python
api_token = "MY_API_TOKEN"
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("stabilityai/stablecode-instruct-alpha-3b", token=api_token)
model = AutoModelForCausalLM.from_pretrained(
  "stabilityai/stablecode-instruct-alpha-3b",
  token=api_token,
  trust_remote_code=True,
  torch_dtype="auto",
)
model.cuda()
inputs = tokenizer("###Instruction\nGenerate a python code that connects to google cloud translate api to translate realtime.###Response\n", return_token_type_ids = False, return_tensors="pt").to("cuda")
tokens = model.generate(
  **inputs,
  max_new_tokens=1000,
  temperature=0.2,
  do_sample=True,
)
print(tokenizer.decode(tokens[0], skip_special_tokens=True))
```
